```python
import IOHMM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
n_components = 2
startprob = np.array([0.2, 0.8])

sd_noise = 0.25
poisson_mean = 2.0
sample_size = 1000
Y = np.random.poisson(poisson_mean, sample_size)
X1 = np.floor(3 * Y + 2 + np.random.normal(0, sd_noise, sample_size))
X2 = np.floor(7 * Y + 5 + np.random.normal(0, sd_noise, sample_size))
X = np.matrix(np.vstack((X1, X2)))
X = X.transpose()

print(X.shape)
print(Y.shape)

hmm = IOHMM._BaseIOHMM(n_components, [X], [Y], start_prob=startprob)

hmm.fit()
```

```python
predictions = hmm._test_predictions()

predictions = np.array(predictions)
predictions = predictions.flatten()
# plot only points not the line
plt.scatter(range(len(Y)), Y, label='True')
plt.scatter(range(len(predictions)), predictions, label='Predicted')
plt.legend()
```
