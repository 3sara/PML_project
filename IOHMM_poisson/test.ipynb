```python
import IOHMM
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
n_components = 2
startprob = np.array([0.2, 0.8])

sd_noise = 0.25
poisson_mean = 2.0
sample_size = 200
Y = np.random.poisson(poisson_mean, sample_size)
X1 = np.floor(3 * Y + 2 + np.random.normal(0, sd_noise, sample_size))
X2 = np.floor(7 * Y + 5 + np.random.normal(0, sd_noise, sample_size))
X = np.matrix(np.vstack((X1, X2)))
X = X.transpose()

print(X.shape)
print(Y.shape)

hmm = IOHMM._BaseIOHMM(n_components, [X], [Y], start_prob=startprob)

hmm.fit()
```
